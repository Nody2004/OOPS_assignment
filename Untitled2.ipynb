```python
#1. What are the five key concepts of Object-Oriented Programming (OOP)?
"""
The five key concepts of Object-Oriented Programming are:-

Classes and Objects:- A class is a blueprint for creating objects, which are instances of the class. Classes define the properties and behaviors of their objects.

Encapsulation:- Encapsulation is the idea of wrapping data and the methods that work on it within a single unit. This limits direct access to variables and methods, which can help prevent accidental data modification

Inheritance:- Inheritance allows attributes and methods to be passed from a base class to a derived class.

Polymorphism:- Polymorphism allows multiple behaviors to be defined for the same class or method.

Abstraction:- Abstraction is the process of hiding the internal complexity of a class while only requiring the necessary data to function.
"""

#2. Write a Python class for a `Car` with attributes for `make`, `model`, and `year`. Include a method to display the car's information.

class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def info(self):
        return f"{self.year} {self.make} {self.model}"

if __name__ == "__main__":
    my_car = Car("Honda", "City", 2023)
    print(my_car.info())


#3. Explain the difference between instance methods and class methods. Provide an example of each.
"""
Instance methods:- These methods are specific to an object, or instance, of a class. They can access both class and instance attributes, and are best for operations on objects. Instance methods require a class instance and can access the instance through the self keyword.
"""
#Example of Instance method:-
class friend:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"{self.name} says: Hello!")

# Example usage
my_friend = friend("Harish")
my_friend.greet()

"""
Class methods:- These methods are related to the class itself, and can access class attributes but not instance attributes. Class methods don't need a class instance, and they have access to the class itself via the cls keyword.
"""
#Example of Class method:-
class Counter:
    count = 0

    def __init__(self):
        Counter.count += 1

    @classmethod
    def get_count(cls):
        return cls.count


counter1 = Counter()
counter2 = Counter()
print(Counter.get_count())


#4. How does Python implement method overloading? Give an example.
"""
Python does not support traditional method overloading like some other languages (e.g., Java or C++).
To achieve behavior similar to method overloading, We can use following technique:
Variable-Length Arguments: Use *args for positional arguments and **kwargs for keyword arguments.
 """
#Example:-
class MathOperations:
    def add(self, *args):
        return sum(args)


math_op = MathOperations()
print(math_op.add(7))
print(math_op.add(5, 10))
print(math_op.add(7, 14, 15, 25))


#5. What are the three types of access modifiers in Python? How are they denoted?.
"""
1. Public Access Modifier:-Members (attributes or methods) defined as public are accessible from outside the class. By default, all members in a Python class are public.
   Denotation: Public members do not have any prefix.

2. Protected Access Modifier:-  Members defined as protected are intended to be accessible only within the class and by subclasses (inherited classes). They are not intended for public use.
   Denotation: Protected members are denoted with a single underscore (_) prefix.

3. Private Access Modifier:- Members defined as private are intended to be accessible only within the class itself. They are not accessible from outside the class or by subclasses.
   Denotation: Private members are denoted with a double underscore (__) prefix.
"""


 #6. Describe the five types of inheritance in Python. Provide a simple example of multiple inheritance.
 """
 There are five types of inheritance in Python:

1. Single Inheritance:
Single inheritance enables a derived class to inherit properties from a single parent class, thus enabling code reusability and the addition of new features to existing code.

2. Multiple Inheritance:
When a class can be derived from more than one base class this type of inheritance is called multiple inheritances. In multiple inheritances, all the features of the base classes are inherited into the derived class.

3. Multilevel Inheritance :
In multilevel inheritance, features of the base class and the derived class are further inherited into the new derived class. This is similar to a relationship representing a child and a grandfather.

4. Hierarchical Inheritance:
When more than one derived class are created from a single base this type of inheritance is called hierarchical inheritance. In this program, we have a parent (base) class and two child (derived) classes.

5. Hybrid Inheritance:
Inheritance consisting of multiple types of inheritance is called hybrid inheritance.
 """
#Example of multiple inheritence:-
class Father:
    def skills(self):
        return "singing, Cooking"

class Mother:
    def skills(self):
        return "Painting, Dancing"

class Child(Father, Mother):
    def show_skills(self):
        father_skills = Father.skills(self)
        mother_skills = Mother.skills(self)
        return f"Father's Skills: {father_skills}, Mother's Skills: {mother_skills}"


child = Child()
print(child.show_skills())


#7. What is the Method Resolution Order (MRO) in Python? How can you retrieve it programmatically?
"""
Method Resolution Order (MRO) in Python refers to the order in which base classes are looked up when searching for a method. This is especially important in the context of multiple inheritance, where a method might be present in more than one parent class. The MRO determines the order in which classes are traversed to find a method or attribute.
The MRO ensures that:

1. Child classes are searched before parent classes.
2. Parent classes are searched in the order they are defined.
3. A class is not considered more than once in the resolution order.

we can retrieve the MRO of a class using the __mro__ attribute or the mro() method.

1. Using the __mro__ attribute: This is a tuple that contains the classes in the order they are searched for methods.

2. Using the mro() method: This method returns a list of classes in the order they are resolved.
 """


#8. Create an abstract base class `Shape` with an abstract method `area()`. Then create two subclasses `Circle` and `Rectangle` that implement the `area()` method.
from abc import ABC, abstractmethod
import math


class Shape(ABC):
    @abstractmethod
    def area(self):
        """Calculate the area of the shape."""
        pass


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        """Calculate the area of the circle."""
        return math.pi * (self.radius ** 2)


class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        """Calculate the area of the rectangle."""
        return self.width * self.height


if __name__ == "__main__":
    circle = Circle(8)
    rectangle = Rectangle(9, 14)

    print(f"Circle area: {circle.area():.2f}")
    print(f"Rectangle area: {rectangle.area():.2f}")



#9. Demonstrate polymorphism by creating a function that can work with different shape objects to calculate and print their areas.
from abc import ABC, abstractmethod
import math


class Shape(ABC):
    @abstractmethod
    def area(self):
        """Calculate the area of the shape."""
        pass


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        """Calculate the area of the circle."""
        return math.pi * (self.radius ** 2)


class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        """Calculate the area of the rectangle."""
        return self.width * self.height


def display_area(shape):
    print(f"Area: {shape.area():.2f}")


if __name__ == "__main__":

    shapes = [
        Circle(6),
        Rectangle(5, 8)
    ]


    for shape in shapes:
        display_area(shape)



#10. Implement encapsulation in a `BankAccount` class with private attributes for `balance` and `account_number`. Include methods for deposit, withdrawal, and balance inquiry.
class BankAccount:
    def __init__(self, account_number, initial_balance=0.0):
        self.__account_number = account_number
        self.__balance = initial_balance

    @property
    def balance(self):
        """Property to access the balance."""
        return self.__balance

    @property
    def account_number(self):
        """Property to access the account number."""
        return self.__account_number

    def deposit(self, amount):
        """Deposit money into the account."""
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount:.2f}. New Balance: {self.__balance:.2f}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        """Withdraw money from the account."""
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew: {amount:.2f}. New Balance: {self.__balance:.2f}")
        else:
            print("Invalid withdrawal amount. Check your balance or the amount.")

# Example usage
if __name__ == "__main__":

    account = BankAccount("812012074", 50000.0)


    account.deposit(17000)
    account.withdraw(20147)
    print(f"Current Balance: {account.balance:.2f}")
    print(f"Account Number: {account.account_number}")



#11. Write a class that overrides the `__str__` and `__add__` magic methods. What will these methods allow you to do?
from math import gcd

class Fraction:
    def __init__(self, numerator, denominator):
        if denominator == 0:
            raise ValueError("Denominator cannot be zero.")
        self.numerator = numerator
        self.denominator = denominator
        self._reduce()

    def _reduce(self):

        common_divisor = gcd(self.numerator, self.denominator)
        self.numerator //= common_divisor
        self.denominator //= common_divisor

    def __str__(self):

        return f"{self.numerator}/{self.denominator}"

    def __add__(self, other):

        if isinstance(other, Fraction):
            new_numerator = (self.numerator * other.denominator) + (other.numerator * self.denominator)
            new_denominator = self.denominator * other.denominator
            return Fraction(new_numerator, new_denominator)  # Return a new reduced fraction
        return NotImplemented

# Example usage
if __name__ == "__main__":
    fraction1 = Fraction(1, 2)
    fraction2 = Fraction(2, 4)

    print(fraction1)
    print(fraction2)

    fraction3 = fraction1 + fraction2
    print(fraction3)



#12. Create a decorator that measures and prints the execution time of a function.
import time
from functools import wraps

def timing_decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        execution_time = end_time - start_time
        print(f"Execution time of '{func.__name__}': {execution_time:.4f} seconds")
        return result
    return wrapper

# Example usage
@timing_decorator
def example_function(n):

    time.sleep(n)
    return f"Waited for {n} seconds."

if __name__ == "__main__":
    result = example_function(2)
    print(result)



#13. Explain the concept of the Diamond Problem in multiple inheritance. How does Python resolve it?
"""
The Diamond Problem occurs in multiple inheritance when a class inherits from two classes that share a common superclass. This creates ambiguity about which superclass method or attribute to use.
Python resolves the Diamond problem by:-
1.Method Resolution Order (MRO): Python uses C3 Linearization to establish a clear order for method and attribute lookup. For a class like D, the MRO determines the search order to avoid ambiguity.

2. Using super(): The super() function can be employed to explicitly call methods from the parent classes according to the MRO, ensuring the correct method is executed.

 """


#14. Write a class method that keeps track of the number of instances created from a class.
class Animal:

    instance_count = 0

    def __init__(self, name):
        self.name = name

        Animal.instance_count += 1

    @classmethod
    def get_instance_count(cls):

        return cls.instance_count

    def __str__(self):

        return f"Animal: {self.name}"

# Example usage
if __name__ == "__main__":
    animal1 = Animal("Dog")
    animal2 = Animal("Cat")
    animal3 = Animal("Rabbit")
    animal4 = Animal("Horse")

    print(animal1)
    print(animal2)
    print(animal3)
    print(animal4)

    print(f"Total number of Animal instances created: {Animal.get_instance_count()}")



#15. Implement a static method in a class that checks if a given year is a leap year.
class LeapYearChecker:
    @staticmethod
    def leap_year(year):
        """Check if a given year is a leap year."""
        return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

# Example usage
if __name__ == "__main__":
    year = 2024
    if LeapYearChecker.leap_year(year):
        print(f"{year} is a leap year.")
    else:
        print(f"{year} is not a leap year.")


    for test_year in [1900, 2000, 2001, 2019, 2022, 2024]:
        print(f"{test_year}: {'Leap year' if LeapYearChecker.leap_year(test_year) else 'Not a leap year'}")






```
